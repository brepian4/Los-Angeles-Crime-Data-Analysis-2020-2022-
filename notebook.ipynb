```python
# Perform one-hot encoding on the 'Vict Descent' column
df = pd.get_dummies(df, columns=['Vict Descent'], prefix='', prefix_sep='')

print(df.head())
```

```python
df = pd.get_dummies(df, columns=['Vict Sex'], prefix='', prefix_sep='')
print(df.head())

```

```python
if 'TIME OCC' in df.columns:
    df['Hour'] = df["TIME OCC"].str.split(':').str[0]
    df['Minute'] = df["TIME OCC"].str.split(':').str[1]
    df['Hour'] = pd.to_numeric(df['Hour'], errors='coerce').fillna(0).astype(int)
    df['Minute'] = pd.to_numeric(df['Minute'], errors='coerce').fillna(0).astype(int)
    df.drop(['TIME OCC'], axis=1, inplace=True)
```

```python
if'DATE OCC' in df. columns:
    df['Year'] = df["DATE OCC"].str.split('/').str[2]
    df['Month'] = df["DATE OCC"].str.split('/').str[0]
    df['Day'] = df["DATE OCC"].str.split('/').str[1]
    df['Year'] = pd.to_numeric(df['Year'], errors='coerce').fillna(0).astype(int)
    df['Month'] = pd.to_numeric(df['Month'], errors='coerce').fillna(0).astype(int)
    df['Day'] = pd.to_numeric(df['Day'], errors='coerce').fillna(0).astype(int)
    df.drop(['DATE OCC'], axis=1, inplace=True)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df['Vict Age'].dropna())
plt.title('Box plot of Vict Age')
plt.ylabel('Vict Age')
plt.show()
```

```python
# Calculate Q1, Q3 and IQR
Q1 = df['Vict Age'].quantile(0.25)
Q3 = df['Vict Age'].quantile(0.75)
IQR = Q3 - Q1

# Define the range for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify the outliers
outliers = df[(df['Vict Age'] < lower_bound) | (df['Vict Age'] > upper_bound)]

# Print the number of outliers
print("Number of outliers in 'Vict Age':", len(outliers))

# If you want to see the outliers
print("Outliers in 'Vict Age':")
print(outliers['Vict Age'])
```

```python
# Import necessary libraries
from scipy import stats
import numpy as np

# Calculate Z-scores for 'Vict Age' column
z_scores = stats.zscore(df['Vict Age'])

# Get a boolean array, True where the z_score value is an outlier
outliers = np.abs(z_scores) > 3

# Print the number of outliers
print("Number of outliers in 'Vict Age':", outliers.sum())

# If you want to see the outliers
print("Outliers in 'Vict Age':")
print(df['Vict Age'][outliers])
```

```python
# Import necessary libraries
import pandas as pd
from scipy import stats
import numpy as np

# Calculate Z-scores
z_scores = pd.DataFrame(stats.zscore(df.select_dtypes(include=[np.number])), columns=df.select_dtypes(include=[np.number]).columns)

# Find outliers
outliers = z_scores[(np.abs(z_scores) > 3).any(axis=1)]

print("Number of outliers:", len(outliers))
```

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number])

# Calculate Q1, Q3 and IQR for each column
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1

# Define the range for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify the outliers
outliers = (numeric_cols < lower_bound) | (numeric_cols > upper_bound)

# Count the number of outliers
num_outliers = outliers.sum().sum()

print("Number of outliers:", num_outliers)
```

```python
# Import necessary libraries
from scipy import stats
import numpy as np

# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number])

# Calculate Z-scores
z_scores = numeric_cols.apply(stats.zscore)

# Identify the outliers
outliers = (np.abs(z_scores) > 3)

# Count the number of outliers
num_outliers = outliers.sum().sum()

print("Number of outliers:", num_outliers)
```

```python
selected_columns = ['AREA', 'Rpt Dist No', 'Vict Age', 'Weapon Used Cd']


corr_matrix = df[selected_columns].corr()


plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, 
                      cbar_kws={'label': 'Correlation coefficient'})

for text in heatmap.texts:
    t = float(text.get_text())
    if 0.5 < np.abs(t) < 0.85:
        text.set_text(text.get_text() + ' *')
        text.set_fontsize(12)
    elif np.abs(t) >= 0.85:
        text.set_text(text.get_text() + ' **')
        text.set_fontsize(14)

plt.title('Correlation Matrix of Selected Columns', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set the style of the plot
sns.set_style("darkgrid")

# Create a palette dictionary
unique_areas = df['AREA NAME'].unique()
palette = sns.color_palette("coolwarm", len(unique_areas))

# Create the plot with a larger size
fig, ax = plt.subplots(figsize=(16,10))

# Create a countplot sorted by count
bar_plot = sns.countplot(y='AREA NAME', data=df, palette=palette, order=df['AREA NAME'].value_counts().index)

# Add a title and increase the font size
plt.title('Count Plot of Area Name', fontsize=30, y=1.02)

# Increase the font size of the x and y labels
plt.xlabel('Count', fontsize=20)
plt.ylabel('Area Name', fontsize=20)

# Increase the font size of the x and y ticks
bar_plot.tick_params(labelsize=15)

# Add annotations
for p in bar_plot.patches:
    width = p.get_width()
    plt.text(5+p.get_width(), p.get_y()+0.55*p.get_height(),
             '{:1.0f}'.format(width),
             ha='center', va='center', fontsize=12, color='black')

# Add a custom grid
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)

# Add a border around the plot
for spine in ax.spines.values():
    spine.set_edgecolor('blue')

# Show the plot
plt.tight_layout()

# Save the plot
plt.savefig('plot.png', dpi=300)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame that counts the number of crimes for each hour in each area
df_grouped = df.groupby(['Hour', 'AREA NAME']).size().reset_index(name='Counts')

# Get the top 5 areas with the most crimes
top_areas = df_grouped.groupby('AREA NAME')['Counts'].sum().nlargest(5).index

# Filter the DataFrame to include only the top areas
df_grouped_top_areas = df_grouped[df_grouped['AREA NAME'].isin(top_areas)]

# Set the style of the plot
sns.set_style("whitegrid")

# Create a line plot
plt.figure(figsize=(15,10))
sns.lineplot(x='Hour', y='Counts', hue='AREA NAME', data=df_grouped_top_areas, palette='viridis', linewidth=2.5)

# Add title and labels
plt.title('Trend of crimes committed hourly in top crime areas', fontsize=20)
plt.xlabel('Hour', fontsize=15)
plt.ylabel('Counts', fontsize=15)

# Show the plot
plt.show()
```
